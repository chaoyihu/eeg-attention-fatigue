```python
# Define function for signal filtering
def filt_coef(order, cutoff, filt_type, sampleRate):
    Nq = sampleRate//2
    Wn = cutoff / Nq
    b, a = scipy.signal.butter(order, Wn, filt_type)
    return b, a
```

```python
################# Multiple subjects #########################
window_length = 1000
datacuts = []
labels = []
threshold = 600
sampleRate = 1000

for fname in newls:
    fpath = os.path.join(r"..\data\data3_attention\raw", fname)
    
#     datacut = np.transpose(np.array(pd.read_csv(fpath)[['FP1', 'FP2']]))

    r = io.read_raw_edf(fpath)
    r.drop_channels(['ECG', 'VEOG', 'HEOG'])
    unfilter_datacut = r.load_data()[['Fp1','Fp2']][0]
    unfilter_datacut = np.array(unfilter_datacut)

    b, a = filt_coef(2, 30, 'lowpass', sampleRate)
    d, c = filt_coef(5, 1, 'highpass', sampleRate)

    unfilter_datacut[0] = scipy.signal.filtfilt(b, a, unfilter_datacut[0])
    unfilter_datacut[1] = scipy.signal.filtfilt(b, a, unfilter_datacut[1])
    unfilter_datacut[0] = scipy.signal.filtfilt(d, c, unfilter_datacut[0])
    unfilter_datacut[1] = scipy.signal.filtfilt(d, c, unfilter_datacut[1])
    
    unfilter_datacut = np.transpose(unfilter_datacut)
    print(unfilter_datacut.shape)
    datacut = []
    for i,j in unfilter_datacut:
        if abs(i) <= threshold and abs(j) <= threshold:
            datacut.append([i,j])
    datacut = np.transpose(np.array(datacut))
    print("datacut:", datacut.shape)

    
    datacut[0] = datacut[0] - min(datacut[0])
    datacut[1] = datacut[1] - min(datacut[1])
    scale = 2 * threshold
    datacut[0] = datacut[0] / scale
    datacut[1] = datacut[1] / scale
    

    datacut = datacut[:, ::2]
    
#     savepath = os.path.join(r"..\data\data2\raw_filtered_baselineSubtracted_thres_fp1fp2_500Hz\raw_filtered_baselineSubtracted_thres_resampled500Hz_fp1fp2_"+ fname.split('.')[0]+'.csv')
#     pd.DataFrame({'FP1':datacut[0], 'FP2':datacut[1]}).to_csv(savepath)

    l = datacut.shape[1] # data points
    n = int(l / window_length) # number of cuts

    # Triple classification: L/M/H
    label = int(fname.split("_")[0][1]) - 1.0 # 0,1,2
    for start in [i * window_length for i in list(range(n))]:
#         label = int(fname.split('_')[-2][1]) - 1
        arr = datacut[:, start: start + window_length]
        datacuts.append(arr)
        labels.append(label)

    print(len(datacuts))
```

```python
################# Multiple subjects #########################
window_length = 1000
datacuts = []
labels = []
threshold = 600
sampleRate = 1000

b, a = filt_coef(2, 30, 'lowpass', sampleRate)
d, c = filt_coef(5, 1, 'highpass', sampleRate)

for fname in newls:
    
    fpath = os.path.join(r"..\data\data3_attention\raw", fname)
    r = io.read_raw_edf(fpath)
    r.drop_channels(['ECG', 'VEOG', 'HEOG'])
    unfilter_datacut = r.load_data()[['Fp1','Fp2']][0]
    unfilter_datacut = np.array(unfilter_datacut)
    
    unfilter_datacut[0] = scipy.signal.filtfilt(b, a, unfilter_datacut[0])
    unfilter_datacut[1] = scipy.signal.filtfilt(b, a, unfilter_datacut[1])
    unfilter_datacut[0] = scipy.signal.filtfilt(d, c, unfilter_datacut[0])
    unfilter_datacut[1] = scipy.signal.filtfilt(d, c, unfilter_datacut[1])
    
    unfilter_datacut = np.transpose(unfilter_datacut)
    print(unfilter_datacut.shape)
    datacut = []
    for i,j in unfilter_datacut:
        if abs(i) <= threshold and abs(j) <= threshold:
            datacut.append([i,j])
    datacut = np.transpose(np.array(datacut))
    print("===========",fname,"=============")
    print("datacut:", datacut.shape)

    
    datacut[0] = datacut[0] - min(datacut[0])
    datacut[1] = datacut[1] - min(datacut[1])
    scale = 2 * threshold
    datacut[0] = datacut[0] / scale
    datacut[1] = datacut[1] / scale

    datacut = datacut[:, ::2]
    
#     savepath = os.path.join(r"..\data\data3_attention\raw_filtered_baselineSubtracted_thres_fp1fp2_500Hz\processed_"+ fname.split('.')[0]+'.csv')
#     pd.DataFrame({'FP1':datacut[0], 'FP2':datacut[1]}).to_csv(savepath)

    l = datacut.shape[1] # data points
    n = int(l / window_length) # number of cuts

#     # Triple classification: L/M/H
#     label = int(fname.split(".")[0][-1]) - 1    # 0,1,2
#     for start in [i * window_length for i in list(range(n))]:
#         arr = datacut[:, start: start + window_length]
#         datacuts.append(arr)
#         labels.append(label)

#     Binary classification: L/H
    label = int(fname.split(".")[0][-1]) - 1    # 0,1,2
    if label == 0:
        for start in [i * window_length for i in list(range(n))]:
            arr = datacut[:, start: start + window_length]
            datacuts.append(arr)
            labels.append(0)
    if label == 2:
        for start in [i * window_length for i in list(range(n))]:
            arr = datacut[:, start: start + window_length]
            datacuts.append(arr)
            labels.append(1)

    print(len(datacuts))
```

### Batch save: 16 fold

```python
def get_group_dataset(newls):
    datacuts = []
    labels = []
    for fname in newls:
        f = open(os.path.join(r"..\data\data3_attention", fname), 'rb')
        data, lbls = pickle.load(f)
        f.close()
        for sample in data:
            datacuts.append(sample)
        for lbl in lbls:
            labels.append(lbl)
    datacuts = np.array(datacuts)
    print(datacuts.shape)
    labels = np.array(labels)
    print(labels.shape)
    
    return datacuts, labels


def save_dataset(datacuts, labels, save_file):
    dataset = [datacuts, labels]
    with open(save_file, 'wb') as handle:
                # pickle.dump(dataset, handle, protocol=pickle.HIGHEST_PROTOCOL)
        pickle.dump(dataset, handle, protocol=2)
        print("saved:", save_file)

subs = ['0002','0004','0005','0006',
        '0007','0009','0013','0201',
        '0401','0502','0601','0701',
        '0801','0901','1101','1301']
```

```python
for test in subs:
    newls = []
    for sub in subs:
        if sub == test:
            continue
        newls.append("attention_"+sub+"_low_fp1fp2_2s_resampled500Hz.pkl")

    datacuts, labels = get_group_dataset(newls)
    save_file = "..\\data\\data3_attention\\attention_f16_"+test+".pkl"
    save_dataset(datacuts, labels, save_file)
```

```python
sampleRate = 500
b, a = filt_coef(2, 30, 'lowpass', sampleRate)
d, c = filt_coef(5, 1, 'highpass', sampleRate)
thres = 100


for fname in newls:
    fpath = os.path.join(r"..\data\data2\raw", fname)
    df = pd.read_csv(fpath)
    fp1 = df['FP1']
    fp2 = df['FP2']
    
    filteredfp1 = scipy.signal.filtfilt(b, a, fp1)
    filteredfp1 = scipy.signal.filtfilt(d, c, filteredfp1)
    filteredfp2 = scipy.signal.filtfilt(b, a, fp2)
    filteredfp2 = scipy.signal.filtfilt(d, c, filteredfp2)
    
    subjects_id = []
    experiences_id = []
    tasks_id = []
    arr1 = []
    arr2 = []

    for idx, pv in enumerate(zip(filteredfp1, filteredfp2)):
        if abs(pv[0]) <= thres and abs(pv[1]) <= thres:
            subjects_id.append(df['subjects_id'][idx])
            experiences_id.append(df['experiences_id'][idx])
            tasks_id.append(df['tasks_id'][idx])
            arr1.append(pv[0])
            arr2.append(pv[1])
    
    arr1 = arr1 - min(arr1)
    arr2 = arr2 - min(arr2)
    scale = 2 * thres
    arr1 = arr1/scale
    arr2 = arr2/scale
    
    df = pd.DataFrame({"subjects_id":subjects_id,
                        "experiences_id":experiences_id,
                        "tasks_id":tasks_id,
                        "FP1":arr1,
                        "FP2":arr2})

    L = df[df['tasks_id']==1]
    M = df[df['tasks_id']==2]
    H = df[df['tasks_id']==3]
    L.to_csv(os.path.join(r"..\data\data2\raw_filtered_baselineSubtracted_thres100", fname.split('.')[0])+'_0.csv')
    M.to_csv(os.path.join(r"..\data\data2\raw_filtered_baselineSubtracted_thres100", fname.split('.')[0])+'_1.csv')
    H.to_csv(os.path.join(r"..\data\data2\raw_filtered_baselineSubtracted_thres100", fname.split('.')[0])+'_2.csv')
```
